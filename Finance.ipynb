## Let's Get Market Data from Different Finances Website like Google Finance, Yahoo Finance and AlphaVantage

We will use Pandas_Datareader for fetching data from Website for tha we need <b>"Stock Name","Finance Name from where we are fetching data [Google Finance]",startdate,enddate</b>.
<br>
For AlphaVantage and Qandl we will need API key to get Data

```python
# Raw Packages
import numpy as np
import pandas as pd
import plotly.graph_objs as go
```

```python
# Data Sources
from pandas_datareader import data,wb
```

```python
# Convert date in right format
# Today's date as variable
from datetime import date
```

```python
startdate = pd.to_datetime('2018-07-15')
enddate = pd.to_datetime(date.today())
```

## Yahoo Finance

```python
data.DataReader("SPY",'yahoo',startdate,enddate)
```

## Google Finance

```python
data.DataReader("AAPL",'stooq',startdate,enddate)
```

## Alphavantage API

```python
# Forex Data
# Conversion from GBP to USD
data.DataReader("GBP/USD",'av-forex',startdate,enddate,api_key='Your api key')
```
